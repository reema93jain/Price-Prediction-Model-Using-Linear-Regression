### 3.1 Linear Regression




In statistical modeling, regression analysis is a set of statistical processes for estimating the relationships between a dependent variable and one or more independent variables. The most common form of regression analysis is linear regression, in which one finds the line that most closely fits the data according to a specific mathematical criterion.

Multiple Linear Regression Equation is given by:   
Y=X1β1 + X2β2 + X3β3...   
where Y is dependent variable i.e. SalePrice   
X1,X2,X3 are indepenedent variables which will help in predicting value for Y   
β1,β2,β3 are the betas   

For the computation of linear regression, I imported module 'linear_model' from python library 'sklearn'. I used below method to predict Y based on X independent variables:   

from sklearn import linear_model           
lr=linear_model.LinearRegression()                 
model=lr.fit(X,Y)                 
predictions=model.predict(X)               
print(f"R² is :{model.score(X,Y)}")                

I checked effectiveness of model using R-squared. R²(The coefficient of determination) is a number between 0 and 1 that measures how well a statistical model predicts an outcome.Generally, a higher coefficient indicates a better fit for the model.

After that I analysis histogram of residuals(i.e. error between actual value of 'Y' & predicted value 'Y') to assess quality of regression model.The Histogram of the Residuals can be used to check whether the variance is normally distributed or not. If the histogram indicates that random error is not normally distributed, it suggests that the model's underlying assumptions may have been violated.       

I also looked at scatter plot for predicted versus actual values of Y. Scatter plots showed straight line in all three models which validates the effectiveness of models.
